## Searching and Sorting

```python
datalist = df.values.tolist()
datalist
```

## Searching

```python
def search(datalist, x):
  for i in range(0, len(datalist)):
    if x in datalist[i][0]:
      print(datalist[i])
      return True
```

```python
search(datalist, "Liver")
```

## Sorting

```python
def bubble_sort(arr):
  arr_length = len(arr)
  for i in range(arr_length-1):
    dey = 0
    for j in range(0, arr_length-i-1):
      if arr[j] > arr [j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
        dey = 1
        if dey == 0:
          break
  return arr
```

```python
sort = (bubble_sort(datalist))
print(sort)
```
